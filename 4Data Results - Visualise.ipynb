# Data Results - Visualisation of Results
---

```python
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.geometry import Point, LineString, Polygon, shape
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
```

```python
## Final store data, grid data, catchment polygons and Lagos LGA Loaded
lagos_grid  =  gpd.read_file("./final_data/lagos_grid.geojson", driver='GeoJSON')
gdf_stores = gpd.read_file("./final_data/lagos_stores.geojson", driver='GeoJSON')
lagos_LGA = gpd.read_file("./data/Lagos_LGA.geojson", driver='GeoJSON')
gdf_catchment = gpd.read_file("./final_data/store_catchment.geojson", driver='GeoJSON')
```

### Grids Visualised

```python
sns.set_theme(style="white")
fig, ax =  plt.subplots(figsize=(25,35))
lagos_LGA.plot(ax=ax, alpha=0.05, edgecolor="black")
lagos_grid.plot(ax=ax, alpha=0.05, edgecolor="black")
```

### Population Visualised

```python
fig, ax =  plt.subplots(figsize=(25,35))
lagos_LGA.plot(ax=ax, alpha=0.05, edgecolor="black")
lagos_grid[lagos_grid["nga_general_2020"]>=10000].plot(ax=ax, figsize=(23,7), legend=True, color='#045a8d')
lagos_grid[lagos_grid["nga_general_2020"]<=10000].plot(ax=ax, figsize=(23,7), legend=True, color='#2b8cbe')
lagos_grid[lagos_grid["nga_general_2020"]<=5000].plot(ax=ax, figsize=(23,7), legend=True, color='#74a9cf')
lagos_grid[lagos_grid["nga_general_2020"]<=1000].plot(ax=ax, figsize=(23,7), legend=True, color='#a6bddb')
lagos_grid[lagos_grid["nga_general_2020"]<=250].plot(ax=ax, figsize=(23,7), legend=True, color='#d0d1e6')
lagos_grid[lagos_grid["nga_general_2020"]<=50].plot(ax=ax, figsize=(23,7), legend=True, color='#f1eef6')
```

### Geocompetition Polygons

```python
fig, ax =  plt.subplots(figsize=(25,35))
lagos_grid.plot(ax=ax, alpha=0.05, edgecolor="black")
lagos_LGA.plot(ax=ax, alpha=0.05, edgecolor="black")
gdf_catchment.plot(alpha=0.08, ax=ax, color="#FF5722")
```

### Competition Level Visualised

```python
fig, ax =  plt.subplots(figsize=(25,35))
lagos_LGA.plot(ax=ax, alpha=0.05, edgecolor="black")
lagos_grid[lagos_grid["competition_level_cat"]=='nan'].plot(ax=ax, figsize=(23,7), legend=True, color='#A6A6A6')
lagos_grid[lagos_grid["competition_level_cat"]=="Very Low"].plot(ax=ax, figsize=(23,7), legend=True, color='#FCBBA1')
lagos_grid[lagos_grid["competition_level_cat"]=="Low"].plot(ax=ax, figsize=(23,7), legend=True, color='#FC9272')
lagos_grid[lagos_grid["competition_level_cat"]=="Medium"].plot(ax=ax, figsize=(23,7), legend=True, color='#FF5722')
lagos_grid[lagos_grid["competition_level_cat"]=="High"].plot(ax=ax, figsize=(23,7), legend=True, color='#DE2D26')
lagos_grid[lagos_grid["competition_level_cat"]=="Very High"].plot(ax=ax, figsize=(23,7), legend=True, color='#A50F15')
```

```python
## Competition Lvl distribution of values
fig, ax = plt.subplots(figsize=(20,8))
sns.histplot(ax=ax, x='competition_level', data=lagos_grid, hue='competition_level_cat', bins=34, palette=['#A6A6A6','#FCBBA1','#FC9272','#FF5722','#DE2D26', '#A50F15'] )
```

```python
## Competition Lvl freq of categories

fig, ax = plt.subplots(figsize=(20,8))
sns.countplot(ax=ax, x="competition_level_cat", data=lagos_grid, palette=['#A6A6A6','#FCBBA1','#FC9272','#FF5722','#DE2D26', '#A50F15'])
```

### Top 10 retailer table

```python
gdf_stores['FloorspaceAVG'] = gdf_stores.Floorspace
```

```python
top10 = gdf_stores.groupby('Company Name').agg({'Price Range': 'first','Store ID' : 'count', 'Floorspace' : 'sum','FloorspaceAVG' : 'mean','Reach' : 'mean','grid_id' : 'sum', 'nga_general_2020' : 'sum'}).sort_values('Floorspace', ascending=False).head(10)
```

```python
top10.nga_general_2020 = top10.nga_general_2020 / 1000000
```
